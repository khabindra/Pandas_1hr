```python
df = pd.DataFrame({"name":['Alfred','Batman','Catwoman'],
                  "toy":[np.nan,'Batmobile','Bullwhip'],
                  "born":[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]})
```

```python
df.head()
```

```python
df.dropna(how='all')
# The code df.dropna(how='all') in pandas is used to remove rows from the DataFrame df where all the values are missing (NaN or NaT). The how='all' parameter specifies that only rows with all missing values should be removed.
```

```python
df2 = pd.DataFrame({
    "name": ['Superman', 'Wonder Woman', 'Aquaman'],
    "toy": [np.nan,np.nan, np.nan],
    "born": [pd.NaT, pd.NaT, pd.Timestamp("1986-11-21")]
})

# NaT stands for "Not a Time" and is used to represent missing or undefined datetime values.
# It is specifically used in datetime columns to denote the absence of a valid datetime value.
# NaT is typically used in columns with data types like datetime64[ns].
```

```python
df2.dropna(how='all',axis=1)
# The dropna method with how='all' and axis=1 removes columns where all values are missing.
# In the provided DataFrame, only the 'toy' column has all missing values (NaN), so that column is removed.
```

```python
df2
# NaN stands for "Not a Number" and is used to represent missing or undefined numerical values.
# It is a floating-point value that can be assigned to any numeric type in pandas.
# NaN is commonly used in numeric columns to denote the absence of a valid numerical value.
# It is typically used in columns with data types like float64.
```

```python
df3 = pd.DataFrame({
    "name": ['Superman', 'Wonder Woman', 'Aquaman'],
    "toy": [pd.NaT, pd.NaT,pd.NaT],
    "born": [pd.NaT, pd.NaT, pd.Timestamp("1986-11-21")]
})
```

```python
df3
```

```python
df3.dropna(how='all',axis=1)
# toy column is removed 
```

```python
df4 = pd.DataFrame({"name":['Alfred','Batman','Alfred'],
                  "toy":[np.nan,'Batmobile','Bullwhip'],
                  "born":[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]})
```

```python
df4.head()
```

```python
df4.drop_duplicates(subset=['name'])
# this removes the duplicates from the column 'name'
```

```python
df4.drop_duplicates(subset=['name'],keep='last')
# removes the duplicates from the column 'name' keeping the last element in the df4
```

```python
df4.drop_duplicates(subset=['name'],keep='first')
# removes the duplicates from the column 'name' keeping the first element
```

```python
df4.drop_duplicates(subset=['name'],keep=False)
# removes all the row which is duplicates ,whose element matches with multiple rows in the dataframe 
```
